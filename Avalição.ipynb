Implementar conceitos de herança e polimorfismo dando continuidade no tema elaborado na prova.

Tema: Agenda 

```python
class Contato: #Criação da classe Contato 
    def __init__(self, nome, telefone, etiqueta):
        self.__nome = nome
        self.__telefone = telefone  #Criar instancia
        self.__etiqueta = etiqueta

    def exibir(self, indice=None):  #Criando metodo exibir
        if indice is not None:
            print(f"\nContato #{indice}")
        print(f"Nome: {self.__nome}")
        print(f"Telefone: {self.__telefone}")
        print(f"Etiqueta: {self.__etiqueta}")

    def tem_etiqueta(self, etiqueta): # criando metod para verificar se tem etiquetas adicionas
        return self.__etiqueta.lower() == etiqueta.lower()

    @property # função para atribuir valores encapsulamento dos dados privados
    def nome(self):
        return self.__nome

    @property
    def telefone(self):
        return self.__telefone


class ContatoProfissional(Contato):  # classe filha que herda de Contato
    def __init__(self, nome, telefone, etiqueta, empresa):
        super().__init__(nome, telefone, etiqueta)
        self.__empresa = empresa

    def exibir(self, indice=None):  # sobrescrevendo o método exibir (polimorfismo)
        super().exibir(indice)
        print(f"Empresa: {self.__empresa}")


class Agenda: #criação da classe agenda, onde vai ser criado todos os metodos da agenda.
    def __init__(self):
        self.contatos = []

    def adicionar_contato(self): # solita os dados para adicionar o contato
        print("1. Contato Pessoal")
        print("2. Contato Profissional")
        tipo = input("Escolha o tipo de contato (1 ou 2): ").strip()

        nome = input("Nome: ").strip()
        telefone = input("Telefone: ").strip()
        etiqueta = input("Etiqueta: ").strip()

        if tipo == '2':
            empresa = input("Empresa: ").strip()
            novo_contato = ContatoProfissional(nome, telefone, etiqueta, empresa)  # cria contato profissional
        else:
            novo_contato = Contato(nome, telefone, etiqueta)  # caso precise adicionar mais contatos

        self.contatos.append(novo_contato)
        print("Contato adicionado!")

    def listar_contato(self): # função usada para listar os contatos que foram adiconadas, exibe elas na tela
        if not self.contatos:
            print(" Agenda vazia.")
            return
        for i, contato in enumerate(self.contatos, start=1):
            contato.exibir(indice=i)

    def buscar_por_etiqueta(self): # funçao que busca por etiquetas, verifica se ja tem alguma existente ou não, exibi se tiver alguma.
        etiqueta = input("Digite a etiqueta: ").strip()
        encontrados = [c for c in self.contatos if c.tem_etiqueta(etiqueta)]

        if not encontrados: # se não houver etiqueta, imprime uma mensagem de erro
            print(" Nenhum contato encontrado com essa etiqueta.")
            return

        print(f"\n Contatos com etiqueta '{etiqueta}':")
        for i, contato in enumerate(encontrados, start=1):
            contato.exibir(indice=i)

    def menu(self): # Menu da agenda, solicitando um numero para cada opção
        while True:
            print("\n Menu da Agenda")
            print("1. Adicionar contato")
            print("2. Listar contatos")
            print("3. Buscar por etiqueta")
            print("4. Sair")

            opcao = input("Escolha uma opção: ").strip()
            if opcao == '1':
                self.adicionar_contato()
            elif opcao == '2':
                self.listar_contato()
            elif opcao == '3':
                self.buscar_por_etiqueta()
            elif opcao == '4':
                print(" Saindo da agenda")
                break
            else:
                print(" Não existe essa opção.") #Caso o usuario inserir alguma opçaõ que nao seja de 1 a 4, sera printado mensagem de erro.


if __name__ == "__main__": #verificação para executar o codigo direto, e nao importado como modulo 
    agenda = Agenda() # criando instacia agenda
    agenda.menu() # chamando o metodo menu da instancia agenda
```
